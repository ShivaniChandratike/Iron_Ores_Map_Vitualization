```python
import folium
import pandas as pd
import numpy as np
```

```python
ds = pd.read_csv('Iron Ore.csv')
latMean = ds['LATDD'].mean()
lngMean = ds['LONDD'].mean()

ironOre = folium.Map(location=[latMean,lngMean],zoom_start=12)

map5 = folium.Map(location=[latMean, lngMean],
                  zoom_start=6, )


def color(elev):
    if elev in range(0, 1000):
        col = 'green'
    elif elev in range(1001, 1999):
        col = 'blue'
    elif elev in range(2000, 2999):
        col = 'orange'
    else:
        col = 'red'
    return col

for lat, lan, name, state in zip(ds['LATDD'], ds['LONDD'], ds['METALLOGENESIS'], ds['STATE']):
    folium.Marker(location=[lat, lan], popup=name,
                  icon=folium.Icon(color=color(state),
                                   icon_color='yellow', icon='cloud')).add_to(map5)
    prev = state
display(map5)    
```
